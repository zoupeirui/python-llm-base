```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# 1. еЈ°жҳҺжҲ‘д»¬иҰҒдҪҝз”Ё Mac зҡ„ GPU (MPS) еҠ йҖҹ
device = torch.device("mps")

# 2. д»Һдә‘з«ҜиҮӘеҠЁдёӢиҪҪе№¶еҠ иҪҪдёҖдёӘдё“дёәиҒҠеӨ©дјҳеҢ–зҡ„е°ҸжЁЎеһӢпјҲзәҰ 1GBпјҢдҪ зҡ„ Mac жҜ«ж— еҺӢеҠӣпјү
model_name = "Qwen/Qwen2.5-0.5B-Instruct"
print("жӯЈеңЁеҠ иҪҪеҲҶиҜҚеҷЁе’ҢжЁЎеһӢпјҢиҜ·зЁҚеҖҷ...")
tokenizer = AutoTokenizer.from_pretrained(model_name)
# to(device) зҡ„ж„ҸжҖқе°ұжҳҜпјҡжҠҠвҖңиҸңи°ұвҖқпјҲжЁЎеһӢпјүзӣҙжҺҘжҗ¬иҝӣ GPU зҡ„ж ёеҝғе·ҘдҪңеҢә
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

# 3. еҮҶеӨҮдҪ зҡ„ PromptпјҲжҸҗзӨәиҜҚпјү
prompt = "иҜ·з”ЁдёҖеҸҘз®Җзҹӯзҡ„иҜқи§ЈйҮҠд»Җд№ҲжҳҜвҖҳз«Ҝдә‘еҚҸеҗҢжҺЁзҗҶвҖҷпјҹ"
messages = [{"role": "user", "content": prompt}]
text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

# 4. е°Ҷж–Үжң¬иҪ¬еҢ–дёә GPU иғҪзңӢжҮӮзҡ„ж•°еӯ—еј йҮҸпјҢе№¶йҖҒе…Ҙ GPU
model_inputs = tokenizer([text], return_tensors="pt").to(device)

# 5. и®© GPU ејҖе§Ӣз–ҜзӢӮи®Ўз®—пјҢз”ҹжҲҗеӣһзӯ”
print("\nеӨ§жЁЎеһӢжӯЈеңЁжҖқиҖғе№¶еӣһзӯ”пјҡ")
generated_ids = model.generate(**model_inputs, max_new_tokens=512)

# 6. жҠҠж•°еӯ—еј йҮҸзҝ»иҜ‘еӣһдәәзұ»иҜӯиЁҖе№¶жү“еҚ°
response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
print(response)
```
